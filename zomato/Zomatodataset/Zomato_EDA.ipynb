```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
df.columns ### so in this case we need tto combine the to dataset beacuse the country code we have in seperate excel file
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how = 'left')
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_values= final_df.Country.value_counts().values
```

```python
## Pie chart
```

```python
plt.pie(country_values,labels=country_name) # in this case the datas are low so its look like dirty
```

```python
## pie chart - Tope 3 countries that uses zomato 
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

## Observation
1. when rating between 4.5 t o4.9 the rating is excellent
2. when rating between 4.0 to 4.4 -------> Very Good
3. when rating between 3.5 to 3.9 ------->good
4. bla bla

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y='Rating Count',data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y='Rating Count',hue='Rating color',data=ratings,palette=["blue","Red","orange",'yellow','green','green'])
```
